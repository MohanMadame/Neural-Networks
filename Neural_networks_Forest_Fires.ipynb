```python
import pandas as pd
import numpy as np
from keras.layers import Dense 
from keras.models import Sequential
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV,KFold
from keras.wrappers.scikit_learn import KerasRegressor      
from tensorflow.keras.optimizers import Adam
```

```python
data= pd.read_csv("/content/forestfires.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
le = LabelEncoder()
data['size_category'] = le.fit_transform(data.size_category)
```

```python
data = data.drop(['month','day'], axis = 1) #as those are strings
```

```python
data.head()
```

```python
x = data.iloc[:,0:28]
y = data.iloc[:,28]
```

```python
std = StandardScaler()
x_scaler = std.fit_transform(x)
```

```python
def create_model():
    model = Sequential()
    model.add(Dense(15,input_dim = 28,kernel_initializer = 'uniform', activation = 'relu'))
    model.add(Dense(10,kernel_initializer = 'uniform', activation = 'relu'))
    model.add(Dense(1,kernel_initializer = 'uniform', activation = 'sigmoid'))
    
    adam = Adam(lr = 0.01)
    model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
    return model
```

```python
#Grid search for all Parameters
model = KerasRegressor(build_fn = create_model, verbose = 0)

batch_size = [20,45,55,70]
epochs = [20,60,70,85]

param_grid = dict(batch_size = batch_size, epochs = epochs)

grid = GridSearchCV(estimator = model, param_grid = param_grid, cv = KFold(), verbose = 10)
grid.fit(x_scaler,y)
```

```python
print('Best: {} with: {}'.format(grid.best_score_ , grid.best_params_))
```

```python
means = grid.cv_results_['mean_test_score']
```

```python
means
```

```python
from keras.layers import Dropout
```

```python
#creating model for dropout and learning rate
def create_model2(dropout_rate,learning_rate):
    model2 = Sequential()
    model2.add(Dense(15,input_dim = 28,kernel_initializer = 'uniform', activation = 'relu'))
    model2.add(Dropout(dropout_rate))
    model2.add(Dense(10,kernel_initializer = 'uniform', activation = 'relu'))
    model2.add(Dropout(dropout_rate))
    model2.add(Dense(1,kernel_initializer = 'uniform', activation = 'sigmoid'))
    
    adam = Adam(lr = 0.01)
    model2.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
    return model2
```

```python
model2 = KerasRegressor(build_fn = create_model2,batch_size = 20, epochs = 60, verbose = 0)

param_grid = dict(dropout_rate = [0.1,0.2,0.5,0.7,0.9], learning_rate = [0.1,0.01,0.001,0.2,0.3])

grid = GridSearchCV(estimator = model2, param_grid = param_grid, cv = KFold(), verbose = 10)
grid.fit(x_scaler,y)
```

```python
print('Best: {} with: {}'.format(grid.best_score_ , grid.best_params_))

mean2 = grid.cv_results_['mean_test_score']
```

```python
def create_model3(activation_function,init):
    model3 = Sequential()
    model3.add(Dense(15,input_dim = 28,kernel_initializer = init, activation = activation_function))
    model3.add(Dropout(0.5))
    model3.add(Dense(10,kernel_initializer = init, activation = activation_function))
    model3.add(Dropout(0.5))
    model3.add(Dense(1,kernel_initializer = 'uniform', activation = 'sigmoid'))
    
    adam = Adam(lr = 0.0001)
    model3.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
    return model3

```

```python
model3 = KerasRegressor(build_fn = create_model3,batch_size = 30, epochs = 70, verbose = 0)

param_grid = dict(activation_function = ['relu','maxout','elu','tanh','sigmoid'], init = ['uniform','normal','zero'])

grid = GridSearchCV(estimator = model3, param_grid = param_grid, cv = KFold(), verbose = 10)
grid.fit(x_scaler,y)
```

```python
print('Best: {} with: {}'.format(grid.best_score_ , grid.best_params_))

```

```python
means3 = grid.cv_results_['mean_test_score']
means3
```

```python
def create_model4(neuron1,neuron2):
    model4 = Sequential()
    model4.add(Dense(neuron1,input_dim = 28,kernel_initializer = 'uniform', activation = 'tanh'))
    model4.add(Dropout(0.5))
    model4.add(Dense(neuron2,kernel_initializer = 'uniform', activation = 'tanh'))
    model4.add(Dropout(0.5))
    model4.add(Dense(1,kernel_initializer = 'uniform', activation = 'sigmoid'))
    
    adam = Adam(lr = 0.0001)
    model4.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
    return model4

```

```python
model4 = KerasRegressor(build_fn = create_model4,batch_size = 30, epochs = 70, verbose = 0)
param_grid = dict(neuron1 = [10,12,15,20], neuron2 = [5,8,10,14])

grid = GridSearchCV(estimator = model4, param_grid = param_grid, cv = KFold(), verbose = 10)
grid.fit(x_scaler,y)
```

```python
print('Best: {} with: {}'.format(grid.best_score_ , grid.best_params_))

```

```python
means4 = grid.cv_results_['mean_test_score']
means4
```

```python
from sklearn.model_selection import cross_val_score
```

```python
def create_model5():
    model5 = Sequential()
    model5.add(Dense(20,input_dim = 28,kernel_initializer = 'uniform', activation = 'tanh'))
    model5.add(Dropout(0.5))
    model5.add(Dense(14,kernel_initializer = 'uniform', activation = 'tanh'))
    model5.add(Dropout(0.5))
    model5.add(Dense(1,kernel_initializer = 'uniform', activation = 'sigmoid'))
    
    adam = Adam(lr = 0.0001)
    model5.compile(loss = 'mean_squared_error', optimizer = 'adam', metrics = ['accuracy'])
    return model5

model5 = KerasRegressor(build_fn = create_model5,batch_size = 30, epochs = 70, verbose = 10)

model5.fit(x_scaler,y)

kfold = KFold(n_splits=30)
results = cross_val_score(model5, x_scaler, y, cv=kfold)
```

```python
print("Results: %.2f (%.2f) MSE" % (results.mean(), results.std()))
```
